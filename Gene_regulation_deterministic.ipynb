```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.integrate import solve_ivp
```

```python
def gen_constitutivo_determinista_modelo(t, r):
    m, p = r
    
    k_m = 1
    g_m = 1/5
    k_p = 50
    g_p = 1/30
    
    dm = k_m - m*g_m
    dp = m*k_p - p*g_p
    
    return np.array([dm, dp])

gen_constitutivo_determinista_dinamica = solve_ivp(gen_constitutivo_determinista_modelo, [0, 200], [0, 0])

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(gen_constitutivo_determinista_dinamica.t, gen_constitutivo_determinista_dinamica.y[0])
plt.title('mRNA')
plt.xlabel('time')
plt.ylabel('concentration')
plt.grid()
plt.subplot(1,2,2)
plt.plot(gen_constitutivo_determinista_dinamica.t, gen_constitutivo_determinista_dinamica.y[1])
plt.title('Protein')
plt.xlabel('time')
plt.ylabel('concentration')
plt.grid()
plt.show()
```

```python
def retroalimentacion_negativa_determinista_modelo(t, r):
    m, p = r
    
    k_r = 1
    K = 5000
    h = 2
    g_m = 1/5
    k_p = 50
    g_p = 1/30
    
    dm = (k_r/(1+(p/K)**h)) - m*g_m
    dp = m*k_p - p*g_p
    
    return np.array([dm, dp])

retroalimentacion_negativa_determinista_dinamica = solve_ivp(retroalimentacion_negativa_determinista_modelo, [0, 200], [0, 0])

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(retroalimentacion_negativa_determinista_dinamica.t, retroalimentacion_negativa_determinista_dinamica.y[0])
plt.title('mRNA')
plt.xlabel('time')
plt.ylabel('concentration')
plt.grid()
plt.subplot(1,2,2)
plt.plot(retroalimentacion_negativa_determinista_dinamica.t, retroalimentacion_negativa_determinista_dinamica.y[1])
plt.title('Protein')
plt.xlabel('time')
plt.ylabel('concentration')
plt.grid()
plt.show()
```
